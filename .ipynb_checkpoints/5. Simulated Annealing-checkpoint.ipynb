When setting T = 1, you recover the Metropolis sampling algorithm.
___
With T=1, you would get a probability of making a move of 50%. This is much smaller than the original 93%, and therefore you would have less of a chance of escaping a maximum. In more intuitive terms, the bouncing ball would still be bouncing, just a lot lower than before.

As you continues to cool the temperature, the ball would bounce less and less. At T=0.1, you would have a probability of making a move of approximately 1%. This means that you wouldn't be able to escape a maximum except for in the luckiest cases. This would be towards the end of the temperature schedule, where you don't want to be bouncing around as much since you've most likely fallen into a maximum.

```python
def function1(x, y):
    return np.exp(-(x**2+ y**2))

def function2(x, y):
    return function1(x,y) + 2*np.exp(-np.abs((x-1.7)**2+(y-1.7)**2))

def function3(x, y):
    return (1-x)**2 + 100*(y-x**2)**2

x = np.linspace(-4, 4)
xx, yy = np.meshgrid(x, x)

fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(xx, yy, function1(xx, yy), alpha=0.5, cmap=cm.Wistia)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$f(x)$')
ax.set_title('Function 1 (eqn. 35)', fontsize=20)

ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(xx, yy, function2(xx, yy), alpha=0.5, cmap=cm.Wistia)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$f(x)$')
ax.set_title('Function 2 (eqn. 36)', fontsize=20)

ax = fig.add_subplot(133, projection='3d')
ax.plot_surface(xx, yy, function3(xx, yy), alpha=0.5, cmap=cm.Wistia)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$f(x)$')
ax.set_title('Function 3 (eqn. 37)', fontsize=20);
```
